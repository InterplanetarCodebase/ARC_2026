# Stone Detection Training (YOLOv26L + CUDA)

This notebook trains a **YOLOv12L** model using Ultralytics on the `stone` dataset for **100 epochs** with **imgsz=640**.

```python
# Install/upgrade required packages for Ultralytics training
%pip install -q --upgrade pip ultralytics
```

```python
from pathlib import Path
import yaml
import torch
from ultralytics import YOLO

print('Torch version:', torch.__version__)
print('Ultralytics imported successfully')
```

```python
# CUDA availability check
cuda_available = torch.cuda.is_available()
print('CUDA available:', cuda_available)
if cuda_available:
    print('GPU:', torch.cuda.get_device_name(0))
    device = 0
else:
    device = 'cpu'
    print('WARNING: CUDA not found, training will run on CPU')
```

```python
# Dataset YAML path
data_yaml = Path('path/data.yaml').resolve()
print('Using data.yaml at:', data_yaml)

with open(data_yaml, 'r', encoding='utf-8') as f:
    data_cfg = yaml.safe_load(f)

print('Classes (nc):', data_cfg.get('nc'))
print('Class names:', data_cfg.get('names'))
print('Train path:', data_cfg.get('train'))
print('Val path:', data_cfg.get('val'))
```

```python
# Load YOLOv12L-seg pretrained weights and train
model = YOLO('yolo26l-seg.pt') # <------------------------ Model chosen as per need

results = model.train(
    data=str(data_yaml),
    epochs=100,
    imgsz=640,
    patience=20,
    device=device,
    project='runs/train',
    name='stone_yolo12l', 
    exist_ok=True
)

results
```

```python
# Validate trained model
metrics = model.val(data=str(data_yaml), imgsz=640, device=device)
metrics
```

```python
# Run sample predictions on test images
test_images = Path('path/test/images')
pred = model.predict(source=str(test_images), imgsz=640, device=device, save=True)
print('Prediction outputs are saved under runs/detect/')
```
